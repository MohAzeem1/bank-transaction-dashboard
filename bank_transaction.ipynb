```python
df['TransactionDate'] = pd.to_datetime(df['TransactionDate'], errors = 'coerce')
```

```python
df['Month'] = df['TransactionDate'].dt.month
```

## Key Insights from Bank Transaction Analysis

### 1.Debit transactions dominate the dataset compared to credit transactions.

### 2.Most transactions involve small amounts, indicating frequent low-value payments.

### 3.Customers use ATM, Online, and Branch channels almost equally.

### 4.Transaction amounts do not show a strong correlation with customer age.

### 5.Customer age has a moderate positive relationship with account balance.

### 6.Transaction activity is evenly distributed across major locations.

### 7.Students perform the highest number of transactions among occupations.

### 8.Transaction activity peaks around October and is lowest around April.
### Most customers maintain low to medium account balances, while very high balances are relatively less common.
### The majority of transactions occur with only one login attempt, indicating normal user behavior and very few suspicious login activities.
### Most transactions are completed within a moderate duration range, suggesting efficient transaction processing with only a few unusually long transactions.
